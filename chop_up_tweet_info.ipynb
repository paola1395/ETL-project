```python
import pandas as pd
```

```python
df = pd.read_csv('tweet_info.csv')
df
```

```python
def df_by_co(myList):
    list_of_dfs = []
    for company in myList:
        cotweet_df = df.loc[df['ticker_symbol'] == company, :]
        list_of_dfs.append(cotweet_df)
    return list_of_dfs
```

```python
df_list = df_by_co(['AAPL','GOOG','GOOGL','AMZN','MSFT', 'TSLA'])
df_list
```

```python
def df_to_csv(myList, company):
    i = 0
    for df in myList:
        df.to_csv(f'./tweet_by_co/tweet_info{company[i]}.csv')
        i+=1
```

```python
df_to_csv(df_list,['AAPL','GOOG','GOOGL','AMZN','MSFT', 'TSLA'])
```
