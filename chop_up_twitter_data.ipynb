```python
import pandas as pd
from datetime import datetime
```

```python
df = pd.read_csv('tweet_data.csv')
df
```

```python
def time_converter(time):
    ts = int(time)
    return str(datetime.utcfromtimestamp(ts).strftime('%Y'))
    
```

```python
df['post_date'] =df['post_date'].map(time_converter)
```

```python
df
```

```python
def df_by_year(myList):
    list_of_dfs = []
    for year in myList:
        my_df = df.loc[df['post_date'] == year, :]
        list_of_dfs.append(my_df)
    return list_of_dfs
```

```python
df_list = df_by_year(['2015','2016','2017','2018','2019'])
df_list
```

```python
def df_to_csv(myList, years):
    i = 0
    for df in myList:
        df.to_csv(f'./tweet_data_by_year/tweet_data{years[i]}.csv')
        i+=1
```

```python
df_to_csv(df_list,['2015','2016','2017','2018','2019'])
```
